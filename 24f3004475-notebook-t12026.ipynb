# Loading datasets

```python
df= pd.read_csv('/kaggle/input/comment-category-prediction-challenge/train.csv')   #train dataset
dt= pd.read_csv('/kaggle/input/comment-category-prediction-challenge/test.csv')    #test dataset
df.head()
```

# Exploratory Data Analysis

```python
X = df.drop(columns=['label']) 
y = df['label']

df.isnull().sum()   # null values present
df.duplicated().sum()  # no duplicates
num= X.select_dtypes(include= 'number')   #numerical columns
cat= X.select_dtypes(include= ['object','boolean'])   #categorical columns
q1= num.quantile(0.25)
q3= num.quantile(0.75)
iqr= q3-q1
lower= q1- 1.5*iqr
upper= q3+ 1.5*iqr
outliers= ((num<=lower)|(num>=upper)).any(axis=1)
outliers.value_counts()    # 35% outliers present
X[num.columns] = num.clip(lower=lower, upper=upper, axis=1)  #Winsorization to remove outliers instead of dropping them
```

# Visualising data

```python
import seaborn as sns
import matplotlib.pyplot as plt

#heatmap
plt.figure(figsize=(8,6))
sns.heatmap(num.corr(), annot= True, cmap= 'coolwarm')
plt.title("Correlation Heatmap")
plt.show()

#boxplot showing no outliers present after winsorization
plt.figure(figsize=(5,3))
for col in ['post_id','upvote','downvote','if_1','if_2']:
    sns.boxplot(x=X[col])
    plt.title(col)
    plt.show()

#histogram
plt.figure(figsize=(6,4))
sns.histplot(df['post_id'], bins=20, kde=True)
plt.title("Distribution of post_id")
plt.show()
```

# Imputing missing values

```python
from sklearn.impute import SimpleImputer

si_n= SimpleImputer(missing_values= np.nan, strategy='mean')  #using mean to impute numerical columns
si= SimpleImputer(missing_values= np.nan, strategy='most_frequent')  #using most frequent to impute categorical columns
X[num.columns]=si_n.fit_transform(X[num.columns])
X[cat.columns]=si.fit_transform(X[cat.columns])

dt[num.columns]=si_n.transform(dt[num.columns])
dt[cat.columns]=si.transform(dt[cat.columns])
print(X.isna().sum().sum())
```

# Feature engineering

```python
X['disability'] = X['disability'].astype(int)
X['created_date'] = pd.to_datetime(X['created_date'])
X['year'] = X['created_date'].dt.year
X['month'] = X['created_date'].dt.month
X['day'] = X['created_date'].dt.day
X['weekday'] = X['created_date'].dt.weekday
X = X.drop(columns=['created_date'])

dt['disability'] = dt['disability'].astype(int)
dt['created_date'] = pd.to_datetime(dt['created_date'])
dt['year'] = dt['created_date'].dt.year
dt['month'] = dt['created_date'].dt.month
dt['day'] = dt['created_date'].dt.day
dt['weekday'] = dt['created_date'].dt.weekday
dt = dt.drop(columns=['created_date'])
```

# Encoding categorical features

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(X[['race','religion','gender']])
X_encoded = pd.DataFrame(ohe.transform(X[['race','religion','gender']]).toarray(), columns=ohe.get_feature_names_out(['race','religion','gender']),index=X.index)
dt_encoded = pd.DataFrame(ohe.transform(dt[['race','religion','gender']]).toarray(), columns=ohe.get_feature_names_out(['race','religion','gender']),index=dt.index)

X = X.drop(['race','religion','gender'], axis=1).join(X_encoded)
dt = dt.drop(['race','religion','gender'], axis=1).join(dt_encoded)
X.head(2)
```

# Scaling numeric features

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# ss= StandardScaler()
# X[num.columns]= ss.fit_transform(X[num.columns])
# dt[num.columns]= ss.transform(dt[num.columns])

pipe_ss = Pipeline([('scaler', StandardScaler())])   # used pipeline for this step
X[num.columns]= pipe_ss.fit_transform(X[num.columns])
dt[num.columns]= pipe_ss.transform(dt[num.columns])

X.head(2)
```

# Text cleaning using regex

```python
import re
def clean(text):
    text = re.sub(r'http\S+|www\S+', '', text)   #any web links replaced with space
    text = re.sub(r'\d+', '', text)     #any digits replaced 
    text = re.sub(r'\S+@\S+\.\S+', '', text)  #any emails replaced
    text = re.sub(r'[^\w\s]', '', text)   #any other charecters excluding words and spaces
    text = text.lower().strip()        
    return text

X['comment']= X['comment'].apply(clean)
dt['comment']= dt['comment'].apply(clean)
X['comment']
```

# Text feature extraction 

```python
from sklearn.model_selection import train_test_split

X_tr, X_t, y_tr, y_t =train_test_split(X,y, test_size=0.25, random_state=42,stratify=y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=7000, min_df=5, ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(X_tr['comment'])
X_train_tfidf = tfidf.fit_transform(X_tr['comment'])
X_test_tfidf = tfidf.transform(X_t['comment'])
dt_tfidf= tfidf.transform(dt['comment'])

X_train_num = X_tr.drop(columns=['comment'])
X_test_num = X_t.drop(columns=['comment'])
dt_= dt.drop(columns=['comment'])
```

```python
from scipy import sparse
from scipy.sparse import hstack
X_tr_num_sparse = sparse.csr_matrix(X_train_num.values)
X_t_num_sparse = sparse.csr_matrix(X_test_num.values)
dt_sparse= sparse.csr_matrix(dt_.values)

X_tr_final = hstack([X_tr_num_sparse, X_train_tfidf])
X_t_final = hstack([X_t_num_sparse, X_test_tfidf])
dt_final= hstack([dt_sparse, dt_tfidf])
```

# Model training, model performance and Hyperparameter tuning

1) Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

lr = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='saga')
lr.fit(X_tr_final, y_tr)
y_pred_lr= lr.predict(X_t_final)

print("Accuracy:", accuracy_score(y_t, y_pred_lr))
print("Classification Report:\n", classification_report(y_t, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_t, y_pred_lr))
```

```python
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV

sk = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
p_lr = {'C': [0.5, 2],     
        'solver': ['saga']}
hpt1 = RandomizedSearchCV(estimator=lr, param_distributions=p_lr, n_iter=8,scoring='f1_macro',cv=sk, n_jobs=-1,random_state=42,verbose=1)
hpt1.fit(X_tr_final, y_tr)

print("Best Parameters:", hpt1.best_params_)
print("Best Score :", hpt1.best_score_)
```

2. LightGBM

```python
from lightgbm import LGBMClassifier

lgb = LGBMClassifier(objective='multiclass',num_class=len(set(y_tr)), n_estimators=500,random_state=42)
lgb.fit(X_tr_final,y_tr)
y_pred_lgb= lgb.predict(X_t_final)

print("Accuracy:", accuracy_score(y_t, y_pred_lgb))
print("Classification Report:\n", classification_report(y_t, y_pred_lgb))
print("Confusion Matrix:\n", confusion_matrix(y_t, y_pred_lgb))
```

```python
sk = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
p_lgb= {'n_estimators': [300,600],  #[300,600,800]
              'max_depth': [-1,7],  
              'num_leaves': [64],  #[31,64]
              'learning_rate': [0.03],    #[0.03, 0.05]
            'min_child_samples': [50], #[20,50]
           'feature_fraction': [0.9]}  #[0.7,0.9]
hpt2 = RandomizedSearchCV(estimator=lgb, param_distributions=p_lgb,n_iter= 8,scoring='f1_macro',cv=sk, verbose=1, n_jobs=-1)
hpt2.fit(X_tr_final, y_tr)

print("Best Parameters:", hpt2.best_params_)
print("Best Score :", hpt2.best_score_)
```

3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_tr_final, y_tr)
y_pred_rf = rf.predict(X_t_final)

print("Accuracy:", accuracy_score(y_t, y_pred_rf))
print("Classification Report:\n", classification_report(y_t, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_t, y_pred_rf))
```

```python
sk = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
p_rf = {'n_estimators': [100,300],
             'max_depth': [None, 3]}
hpt3 = RandomizedSearchCV(estimator=rf, param_distributions=p_rf,n_iter= 8,scoring='f1_macro',cv=sk, verbose=1, n_jobs=-1)
hpt3.fit(X_tr_final, y_tr)

print("Best Parameters:", hpt3.best_params_)
print("Best Score:", hpt3.best_score_)
```

# Final predication

```python
# Predicting on test dataset with highest score model - Lightgbm 0.802
final_lgb = hpt2.best_estimator_
y_pred = final_lgb.predict(dt_final)
```
